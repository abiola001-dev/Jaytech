üßπ Step 9: Handle Missing Values

```python
# Impute missing values with median of each column
df.fillna(df.median(), inplace=True)

# Confirm no missing values remain
df.isnull().sum()
```

‚öñÔ∏è Step 10: Feature Scaling with StandardScaler
Machine learning models often benefit from scaling. We'll use StandardScaler from sklearn.

```python
from sklearn.preprocessing import StandardScaler

# Separate features and target
X = df.drop('Outcome', axis=1)
y = df['Outcome']

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

üß† Real-World Analogy
Imagine you‚Äôre grading students, but one test is out of 100 and another is out of 500. You can't compare them directly. You‚Äôd want to normalize the scores (e.g., convert everything to percent or z-scores). That‚Äôs what StandardScaler does ‚Äî it makes all features comparable.

üîÄ Step 11: Train-Test Split
We split the data into training and testing sets using train_test_split.

```python
from sklearn.model_selection import train_test_split

# Split the scaled data into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.4, random_state=42, stratify=y)
```

üîç What We're Doing
We want to train a model on part of the data and then test it on unseen data to evaluate how well it performs. This means we need to split our dataset into:

Training set ‚Äî used to teach the model.

Testing set ‚Äî used to check if the model actually learned something meaningful (not just memorized the training data).

üß† Understanding the Code
from sklearn.model_selection import train_test_split

üì¶ Imports the function train_test_split from Scikit-learn.

This function randomly splits your data into training and testing subsets.

| Component         | Meaning                                                                                                                                            |
| ----------------- | -------------------------------------------------------------------------------------------------------------------------------------------------- |
| `X_scaled`        | These are your **features** (i.e., inputs like glucose, BMI, etc.).                                                                                |

| `stratify=y`      | Makes sure the **distribution of diabetic vs. non-diabetic** is the same in both the training and test sets. This prevents class imbalance issues. |


| Variable  | Description                                           |
| --------- | ----------------------------------------------------- |
| `X_train` | Feature data for the training set (80% of `X_scaled`) |
| `X_test`  | Feature data for the testing set (20% of `X_scaled`)  |
| `y_train` | Target labels (0 or 1) for the training set           |
| `y_test`  | Target labels for the testing set                     |

ü§ñ Step 12: Train a Basic Model (Logistic Regression)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Train the model
model = LogisticRegression()
model.fit(X_train, y_train)
```

üîç Step-by-Step Explanation:
üîπ model = LogisticRegression()
You're creating (instantiating) a machine learning model: Logistic Regression

Logistic Regression is a popular classification algorithm ‚Äî perfect for binary outcomes (like predicting diabetes: 0 = no, 1 = yes)

At this point, the model exists, but it hasn't learned anything yet ‚Äî it‚Äôs like an empty brain üß†

Think of this line like saying: ‚ÄúHey scikit-learn, I want to use logistic regression.‚Äù

üîπ model.fit(X_train, y_train)
Now you‚Äôre training the model using:

X_train: your 614 rows of training features

y_train: your 614 labels (the correct outcomes: 0 or 1)

This is where the model learns the relationship between features (like Glucose, BMI, Age) and the outcome (diabetic or not).

Behind the scenes:

Logistic Regression calculates coefficients (weights) for each feature

It tries to find the best fit that minimizes error between predicted and actual labels

```python
# Predict on test set
y_pred = model.predict(X_test)
```

What this line does:
model.predict(X_test) uses the trained Logistic Regression model to predict the output labels for your test features X_test.

For each row in X_test, it applies the learned formula (with weights and bias) and passes the result through the sigmoid function.

If the predicted probability is > 0.5, it assigns a label 1 (positive class, e.g., diabetic).

If the predicted probability is ‚â§ 0.5, it assigns a label 0 (negative class, e.g., non-diabetic).

What you get in y_pred:
A NumPy array or pandas Series of predicted labels (0s and 1s) for each test sample.

Same number of rows as X_test.

```python
# Evaluate performance
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

What does the confusion matrix represent?

Based on your earlier report, it should look something like this:

Predicted    0	      Predicted 1
Actual 0	~81	        ~19
Actual 1	~27	        ~27

True Negatives (TN) = correctly predicted non-diabetics (0)

False Positives (FP) = non-diabetics predicted as diabetics (Type I error)

False Negatives (FN) = diabetics predicted as non-diabetics (Type II error)

True Positives (TP) = correctly predicted diabetics (1)

What does each cell mean?
True Negatives (TN) = 81
These are the samples that were actually non-diabetic (0) and the model correctly predicted them as non-diabetic. This is good ‚Äî the model did not raise false alarms here.

False Positives (FP) = 19
These samples were actually non-diabetic (0) but the model incorrectly predicted them as diabetic (1). This is a Type I error ‚Äî false alarm, predicting diabetes when there isn‚Äôt any.

False Negatives (FN) = 27
These samples were actually diabetic (1) but the model predicted them as non-diabetic (0). This is a Type II error ‚Äî missing true diabetic cases, which is usually the most concerning error in medical diagnosis.

True Positives (TP) = 27
These are the samples that were actually diabetic (1) and the model correctly predicted them as diabetic.

What insights can you draw?
The model is doing fairly well on non-diabetics ‚Äî it correctly identifies 81 out of 100.

It makes more mistakes on diabetic cases: it misses 27 diabetics (false negatives), which is almost half of the diabetic cases (54 total).

False negatives are a concern in healthcare because these patients might miss timely treatment.

What to do next?
Focus on reducing False Negatives (FN) to catch more true diabetics ‚Äî you could:

Adjust the decision threshold (e.g., lower from 0.5 to increase sensitivity)

Use class weights in logistic regression to penalize FN more

Try other models or ensemble methods

Use techniques like SMOTE to balance the classes

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.metrics import (
    classification_report, confusion_matrix, accuracy_score,
    roc_curve, auc
)

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y)

# 1. Logistic Regression
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
log_pred = log_model.predict(X_test)
log_proba = log_model.predict_proba(X_test)[:, 1]

# 2. XGBoost Classifier
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X_train, y_train)
xgb_pred = xgb_model.predict(X_test)
xgb_proba = xgb_model.predict_proba(X_test)[:, 1]

# --- Evaluation Reports ---
print("Logistic Regression Accuracy:", accuracy_score(y_test, log_pred))
print("XGBoost Accuracy:", accuracy_score(y_test, xgb_pred))
print("\nLogistic Regression Report:\n", classification_report(y_test, log_pred))
print("\nXGBoost Report:\n", classification_report(y_test, xgb_pred))

# --- Confusion Matrix for XGBoost ---
conf_mat = confusion_matrix(y_test, xgb_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', xticklabels=[0, 1], yticklabels=[0, 1])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('XGBoost Confusion Matrix')
plt.show()

# --- ROC Curve & AUC ---
fpr_log, tpr_log, _ = roc_curve(y_test, log_proba)
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, xgb_proba)
auc_log = auc(fpr_log, tpr_log)
auc_xgb = auc(fpr_xgb, tpr_xgb)

plt.figure(figsize=(10, 6))
plt.plot(fpr_log, tpr_log, color='blue', lw=2, label=f'Logistic Regression (AUC = {auc_log:.2f})')
plt.plot(fpr_xgb, tpr_xgb, color='green', lw=2, label=f'XGBoost (AUC = {auc_xgb:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison')
plt.legend(loc='lower right')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import auc, roc_curve
import matplotlib.pyplot as plt

# Calculate ROC curve and AUC
fpr_log, tpr_log, _ = roc_curve(y_test, log_proba)
auc_log = auc(fpr_log, tpr_log)

fpr_xgb, tpr_xgb, _ = roc_curve(y_test, xgb_proba)
auc_xgb = auc(fpr_xgb, tpr_xgb)

plt.figure(figsize=(10, 6))

# Plot ROC curves
plt.plot(fpr_log, tpr_log, color='blue', lw=2, label='Logistic Regression')
plt.plot(fpr_xgb, tpr_xgb, color='green', lw=2, label='XGBoost')

# Fill area under the curve (shading)
plt.fill_between(fpr_log, tpr_log, alpha=0.2, color='blue', label=f'LogReg AUC = {auc_log:.2f}')
plt.fill_between(fpr_xgb, tpr_xgb, alpha=0.2, color='green', label=f'XGBoost AUC = {auc_xgb:.2f}')

# Random guess line
plt.plot([0, 1], [0, 1], color='red', linestyle='--', lw=1.5, label='Random Guess')

# Plot formatting
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve with AUC Area Shading')
plt.legend(loc='lower right')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

üß≠ Axes:
X-axis: False Positive Rate (FPR)

The proportion of actual negatives that were wrongly predicted as positive

Lower is better

Y-axis: True Positive Rate (TPR) (aka Sensitivity or Recall)

The proportion of actual positives correctly predicted

Higher is better
